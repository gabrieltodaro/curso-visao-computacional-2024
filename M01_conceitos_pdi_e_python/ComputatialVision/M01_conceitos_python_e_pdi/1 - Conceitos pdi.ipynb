# 1 - Conceitos PDI

```python
import numpy # Math library
import matplotlib.pyplot as plt # Chart library.
from PIL import Image

img = Image.open('./cameraman/cameraman.tiff')
img = numpy.array(img)
img.shape
plt.imshow(img)
```

```python
# plt.imshow(img, cmap='gray')
plt.figure(figsize = (12,4))
_ = plt.hist(img.flatten(), bins = 256)
```

```python
img_transformed = img.astype(float) ** 4 # Saturando a imagem
plt.imshow(img_transformed, 'gray')
print(img_transformed.min(), img_transformed.max())
```

```python
img_transformed = img.astype(float) ** 0.2 # Clareando a imagem
plt.imshow(img_transformed, 'gray')
```

```python
# Normalizando a imagem: levando o mínimo para 0 e o máximo para 255
img_normalized = img_transformed - img_transformed.min()
img_normalized = 255 * img_normalized / img_normalized.max()
print(img_normalized.min(), img_normalized.max())

img_normalized = img_normalized.astype(numpy.uint8)
_ = plt.hist(img_normalized.flatten(), bins = 256)
```

## Formato de imagens

```python
root = './cameraman/'
img_or = Image.open(root+'cameraman.tiff')
image2 = Image.open(root+'cameraman2.tiff')
image3 = Image.open(root+'cameraman3.tiff')
image4 = Image.open(root+'cameraman4.tiff')
image_jpeg = Image.open(root+'cameraman.jpeg')
image_png = Image.open(root+'cameraman.png')

images = [img_or, image2, image3, image4, image_jpeg, image_png]
for image in images:
    plt.figure()
    plt.imshow(image, cmap = 'gray')

def info(image):
    image = numpy.array(image)
    print('dtype:', image.dtype)
    print('shape:', image.shape)
    print('values:', image.min(), image.max())
    print('\n')

for image in images:
    info(image)
```
